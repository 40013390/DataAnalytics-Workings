```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Service_Requests_From_2010.csv")
```

```python
df
```

```python
city_group=df.groupby("City")
```

```python
city_group.groups.keys()
```

```python
Complaint_group=df.groupby("Complaint Type")
```

```python
for city,df in city_group:
    print("{} , {}\n".format(city,len(df)))
```

```python
Complaint_list=[]
Location_list=[]
for loc,df in city_group:
    print("{} ,{}".format(loc,len(df),axis=0))
    Location_list.append(loc)
    Complaint_list.append(len(df))
```

```python
df=pd.DataFrame(
    {
        "City":Location_list,
        "Complaint_count":Complaint_list
    }
)
```

```python
import matplotlib.pyplot as plt
list1=[]
list2=[]
for City,d in city_group:
    list1.append(City)
    list2.append(len(d))
    
plt.scatter(list1,list2,marker="*",color="red")
plt.xlabel=("Cities")
plt.ylabel=("complaints ")
plt.show()
```
